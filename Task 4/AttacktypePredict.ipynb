```python
import numpy as np
import xlrd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def read_excel(file,sheet):
    ExcelFile = xlrd.open_workbook(file)
    #print(ExcelFile.sheet_names())
    sheet = ExcelFile.sheet_by_name(sheet)
    nrows = sheet.nrows
    ncols = sheet.ncols
    X = np.zeros((nrows,ncols-1))
    Y = np.zeros((nrows,1))
    for j in range(ncols-1):
        cols = sheet.col_values(j)
        X[:,j] = cols
    Y[:,0] = sheet.col_values(ncols-1)

    return X,Y
```

```python
trainData,trainLabel = read_excel('q4dataset.xlsx','train_v1')
testData,testLabel = read_excel('q4dataset.xlsx','test_v1')
print('trainData.shape = ' ,trainData.shape)
print('trainLebal.shape = ',trainLabel.shape)
print('testData.shape = ', testData.shape)
print('testLebal.shape = ',testLabel.shape)
```

```python
def classify(inX, dataSet, labels, k = 5):
    '''
        inX.shape = (1,5)
        dataSet.shape = (100000, 5)
        labels = (1,1)
        
    ''' 
    #转成mat
    inX = np.mat(inX)
    dataSet = np.mat(dataSet)
    labels = np.mat(labels)
    
    # calc number of sample
    dataSetSize = dataSet.shape[0]
    # create inX with the same number of sample to diff
    # 沿y轴复制两倍，沿x轴复制1倍
    diffMat = np.tile(inX,(dataSetSize,1)) - dataSet
    # corresponding point to square
    sqDiffMat = np.array(diffMat)**2
    #axis = 1 to horizontal sum
    sqDistances = sqDiffMat.sum(axis=1)
    #square root
    distances = sqDistances**0.5
    #args : 返回按照从小到大排序以后，数组中某个值应当排的位置的坐标
    # x = np.array([3, 1, 2])  --  np.argsort(x)  --  array([1, 2, 0])
    sortedDistIndicies = distances.argsort()
    classCount = {}
    for i in range(k):
        voteIlabel = labels[ sortedDistIndicies[i],0]
        classCount[voteIlabel] = classCount.get(voteIlabel, 0) + 1
    sortedClassCount = sorted(classCount.items(), key = lambda item:item[1], reverse=True)
    #print(sortedClassCount)
    #return label
    return sortedClassCount
```
